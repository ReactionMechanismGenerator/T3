## Create a flux diagram

```python
import os
from IPython.display import display
from PIL import Image

from t3.common import DATA_BASE_PATH

import t3.utils.flux as flux
```

##### Define the model path and the folder path

```python
model_path = os.path.join(DATA_BASE_PATH, 'models', 'HOCHO.yaml')

folder_path = os.path.join('HOCHO_flux_diagrams')
```

```python
observables = ['HOCHO(1)', 'H(3)', 'HOCO(10)']
flux.generate_flux(model_path=model_path,
                   folder_path=folder_path,
                   observables=observables,
                   times=[2.0],
                   composition={'HOCHO(1)': 0.2, 'N2': 0.98},
                   T=1200,
                   P=1,
                   reactor_type='JSR',
                   explore_tol=0.95,
                   dead_end_tol=0.10,
                   generate_separate_diagrams_per_observable=True,
                   display_flux_ratio=True,
                   display_concentrations=True,
                   display_r_n_p=True,
                   )
```

```python
for observable in observables:
    img = Image.open(os.path.join(folder_path, 'flux_diagrams', observable, 'flux_diagram_2.0_s.png'))
    print(observable)
    display(img)
    print('\n\n')
```
